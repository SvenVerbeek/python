```python
import pandas as pd
```

```python
import xml.etree.ElementTree as ET
```

```python
from tkinter import Tk
from tkinter.filedialog import askopenfilename
```

```python
import re
```

```python
def checkData(tag):
    try:
        match = re.search('(?:{.*?})(.*)', tag)
        newTag = match.group(1)
    except:
        newTag = tag
    return newTag
```

```python
Tk().withdraw() # we don't want a full GUI, so keep the root window from appearing
file = askopenfilename() # show an "Open" dialog box and return the path to the selected file
```

```python
parser = ET.XMLParser(target=ET.TreeBuilder(insert_comments=True))
```

```python
tree = ET.parse(file, parser)
```

```python
root = tree.getroot()
```

```python
for index, child in enumerate(root[1][0].iter()): 
    checkChildren(child)
    node = checkData(child.tag)
    
    if "function Comment" in str(node):
        nodeComment = child.text[:-1]
        nextHasComment = True
    elif index == 0:
        title = node
        df = pd.DataFrame(columns=[title, 'Annotation', 'Description'])
    else:
        df.loc[len(df)] = {title: node, 'Annotation' : "("+nodeComment+")", 'Description' : ""} \
        if (nextHasComment and index != 0) else {title: node, 'Annotation' : "", 'Description' : ""}
        nextHasComment = False
```
